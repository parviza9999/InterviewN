Conv —> Relu —> Flatten --> Linear --> Output
Convolutional layer - use 32 filters, kernel size 3x3 and Relu activation, Linear 128 (layers), output 10 classes. MNIST dataset

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
# input image 28x28, kernel = 3x3 , 32 kernels ,
class KGCNN(nn.module):
  def __init__(self):
    super(KGCNN, self).__init__()
    self.conv1 = nn.Conv2d(1, 32, 3, 1)
    self.linear1 = nn.Linear(32*26*26,128)
    self.output = nn.Linear(128,10)
  def forward(self, x):
    x = self.conv1(x)
    x = F.relu(x)
    x = nn.Flatten(x)
    x = self.linear1(x)
    x = self.output(x)
    return x
model = KGCNN()
print(model)
```
